# 수행과제 3
힙을 이용하여 생일이 늦은 순으로 10명을 출력한다.

```python
import csv
import heapq

heap = []

with open('birthday.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)

    for row in reader:
        try:
            name = row[0].strip()
            year = int(row[1])
            month = int(row[2])
            day = int(row[3])

            heapq.heappush(heap, (-year, -month, -day, name))
        except:
            continue

print("[생일이 늦은 순서 10명]")
count = 0
while heap and count < 10:
    y, m, d, name = heapq.heappop(heap)
    print(f"{count+1}. {name} - {-y}년 {-m}월 {-d}일")
    count += 1
```

# 수행과제 4
원형 이중 연결리스트를 이용하여 같은 조 친구들의 이름과 생년월일을 출력한다.
CSV에는 조 정보가 없고 색으로만 구분되어 있으므로, 같은 조 친구 이름 목록을 코드에서 직접 정의하였다.
생년월일이 없는 경우는 예외처리하여 건너뛰었다.

```python
import csv
from DS.list.circularDoublyLinkedList import CircularDoublyLinkedList

my_group = ['황다원', '박다인', '안예원', '정서하', '김우현', '양시언', '유가현', '이하연', '장예은', '전은빈', '정세희', '이예은']

cdll = CircularDoublyLinkedList()

with open('birthday.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)

    for row in reader:
        try:
            name = row[0].strip()

            if row[1] == '' or row[2] == '' or row[3] == '':
                continue

            year = int(row[1])
            month = int(row[2])
            day = int(row[3])

            cdll.append((name, year, month, day))

        except:
            continue

print("[같은 조 친구들]")

head = cdll._CircularDoublyLinkedList__head
node = head.next

while node != head:
    name, year, month, day = node.item

    if name in my_group:
        print(f"{name} - {year}년 {month}월 {day}일")

    node = node.next
```

# 04

# 05
간단한 일이다.힙의 마지막 원소는 배열의 맨 끝 원소이므로 배열에서 그냥 제거하면 된다. 이 원소는 가장 마지막 위치의 리프 노드이므로 제거해도 완전 이진 트리 조건이 유지된다. 또한 마지막 원소는 자식이 없으므로 힙 조건도 추가 수선 없이 유지된다. 따라서 맨 마지막 원소 삭제는 보통 **Θ(1)**에 가능하다.
# 06
힙은 만들어질 수 있지만, 본문의 방법보다 더 효율적이지 않다. 본문 방식: 아래쪽에서부터 스며내리기 반복 → Θ(n), 학생 방식: 위쪽에서부터 스며오르기 반복 → 각 원소마다 최대 Θ(log n). 따라서 전체는 Θ(n log n). 즉, 위쪽에서부터 스며오르기를 반복하는 방법은 가능하지만 더 비효율적이다.
# 07
   값이 증가한 원소를 위로 스며오르게(percolate up) 하면 된다.
   과정은 다음과 같다.
	1.	값이 증가한 원소의 위치를 찾는다.
	2.	그 원소를 부모와 비교한다.
	3.	부모보다 값이 크면 부모와 자리를 바꾼다.
	4.	이 과정을 더 이상 부모보다 크지 않을 때까지 반복한다.
	5.	최악의 경우 루트까지 올라갈 수 있으므로 시간은 O(log n)이다.
   즉, 최대 힙에서 어떤 원소의 값이 증가하면 아래로 내릴 필요는 없고 위로 올리면 된다.
